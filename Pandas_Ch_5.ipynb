#### Histogram

```python
plt.hist(x = 'mpg', data = car)
plt.xlabel='mpg'
plt.ylabel='frequency'
plt.title('Mileage Distribution')
```

#### Countplot : to use for mainly discrete data( whole number 3 ,4) : univaraint(only one parameter to pass)

```python
import seaborn as sns
```

```python
sns.countplot(x ='gear', data = car)
plt.figure(figsize= (5,5)) #to increase the size we need matplotlib.pyplot as plt 
```

#### Barplot : bivariant (can pass two values)

```python
plt.bar(x='gear',height='mpg', data=car ) 
```

```python
sns.barplot(x='gear',y='mpg', data=car)
```

```python
plt.scatter(x='hp', y='mpg', data= car)
plt.xlabel='HP'
plt.ylabel='MPG'
```

```python
sns.pairplot(data=car)
```

```python
gear_count= car.gear.value_counts()
#plt.pie(x=gear_count,data=car)
plt.pie(x=gear_count, labels=[3,4,5], explode=[0.1,0.2,0.2])
```

```python
car.gear.value_counts().plot(kind = 'pie', labels=[3,4,5], explode=[0.1,0.2,0.2] )
```

#### Boxplot

```python
plt.boxplot(x='mpg', data=car)
```

```python
sns.boxplot(x='mpg', data=car, color='orange')
```
