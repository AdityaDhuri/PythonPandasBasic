## : Grouping vs Pivottable vs Crosstab

```python
insurance = pd.read_csv("A:/PythonPandas/insurance.csv")
```

```python
insurance.head(10)
```

### : groupby

```python
insurance.groupby(['sex','smoker','region'])['age'].mean()
```

```python
insurance.groupby(['sex'])['smoker'].value_counts()
```

### pivot_table

```python
pd.pivot_table(data = insurance, index = 'smoker', columns = 'region', values = 'charges', aggfunc = 'mean')
```

### : crosstab 

```python
# crossbar displays the frequency distribution of a table
pd.crosstab(index= insurance.region, columns = insurance.smoker, margins = True)
```
